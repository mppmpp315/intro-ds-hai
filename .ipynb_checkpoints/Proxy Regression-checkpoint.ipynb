#### Overview

Main Idea:
Proxy regression is way in which you try to predict target variable Y by training on a proxy variable Y_s.
It is important to not that by doing so we are assuming that the target variable (Y) is highly correlated with the proxy variable (Y_s), which is why trainning on one to predict the other actually works.

This is useful in situations where a binary target variable has very few "positive" occurences and the proxy variable (Y_s) has significantly more "positive" occurances. (According to the professor), it has been shown that many times when a target variable is sparse, training on a proxy and predicting on the target yields much better results than training on the target itself.

In this notebook we will be attempting to create 2 types of proxy regressions:

1) linear regression:

    *attempt to predict the variable Y using Y_s where Y_s is continuous.

2) logistic regression:
    * attempt to predict variable Y using Y_s where Y_s is a binary variable.

**First we must create our Own sklearn.cross_validation.train_test_split.**

This function is needed because:

    a) sklearn.cross_validation.train_test_split takes as input and X and a Y. Since we have 3 elements we want split (X,Y,Y_s), we want to know the indices of the split so we can segment all 3. However, train_test_split returns a ndarrays so we cannot get the indices that way. kfolds ***Does*** return indices, but fails because of the reason describes in b)
    
    
    b) Because the target variable is sparse he have to make sure that we have at least a minimun number of positives in both training and testing data. otherwise we might either end up with a trainning set with all negative entries (which doesn't properly train our model) or end up not having any positives in our testing data (making is impossible to calculates AUC). 

```python
def splitTestTrainIndices(DF , target_column_name, proxy_column_name, min_frac_positive = 0.75):
    """ This function takes in a dataframe and splits it into training and testing dataframes. 
    when splitting the data the function also makes sure at least a fraction of the positive entries (min_frac_positive) are 
    in the training set for both the proxy and target variables.
    
    Args:
       DF: a pandas dataframe on which to do the analysis.
           DF must contain at the very least the columns: [proxy_column_name, target_column_name] and one feature.

       target_column_name: the name of the column containing the binary target variable.
       
       proxy_column_name: the name of the column containing the binary proxy variable.
       
       min_frac_positive: float. the mean fraction of overall positives we want the training DF to contain. default = 0.75.

    """
    boolean = False
    super_set = set(DF.index)
    condition_a = condition_b = condition_c = condition_d = False
    
    #calculate the positives in target and proxy
    min_positives_target = np.floor(np.sum(working_df[target_column_name])*min_frac_positive)
    min_positives_proxy = np.floor(np.sum(working_df[proxy_column_name])*min_frac_positive)
   
    while not condition_a and not condition_b and not condition_c and not condition_d:
        train_ix = np.random.choice(DF.index, size= np.floor(0.8*len(DF.index)), replace=False)
        train_ix = set(train_ix)
        test_ix = super_set-train_ix

        #calculate number of positives in target
        positives_train = DF[target_column_name].ix[train_ix].sum()
        positives_test = DF[target_column_name].ix[test_ix].sum()
        
        #calculate number of positives in proxy
        positives_proxy_train = DF[proxy_column_name].ix[train_ix].sum()
        positives_proxy_test = DF[proxy_column_name].ix[test_ix].sum()
        
        #print positives_train, positives_test, "---", positives_proxy_train, positives_proxy_test
        
        #make sure the is a minimun amount of positives in training
        condition_a = positives_train >= min_positives_target
        condition_b = positives_proxy_train >= min_positives_proxy
        
        #there should be at least 20% of positives in testing, otherwise the AUC cannot be calculated/ will be heavily variable.
        condition_c = positives_test >= np.floor(np.sum(working_df[target_column_name])*0.1)
        condition_d = positives_proxy_test >= np.floor(np.sum(working_df[proxy_column_name])*0.1)
        
    #make sure we didn't lose any indices
    assert (train_ix | test_ix) == super_set
    
    #print positives_train, positives_test, positives_proxy_train, positives_proxy_test
    
    return train_ix, test_ix


splitTestTrainIndices(DF = working_df,target_column_name = 'Bin 2014', proxy_column_name = 'Bin 2013', min_frac_positive = 0.8)
```

## Proxy Logistic Regression

```python

def proxyLogisticRegression(DF,target_column_name, proxy_column_name, min_frac_positive = 0.75):
    """ Performs logostic regression using a proxy variable Y_s. This means that we train the dataset on the proxy variable Y_s
    and test the AUC of our model using the target variable Y. 
    In this function we are also calculating the AUC for Y_s to get a general sense of how good our model is.
    Args:
        DF: a pandas dataframe on which to do the analysis.
           DF must contain at the very least the columns: [proxy_column_name, target_column_name] and one feature.

        target_column_name: the name of the column containing the binary target variable.

        proxy_column_name: the name of the column containing the binary proxy variable.

        min_frac_positive: float. the mean fraction of overall positives we want the training DF to contain. default = 0.75.
        
    Returns:  
        AUC_target: AUC score for target variable Y when trained on Y_s
        AUC_proxy: AUC score for proxy variable Y_s when trained on Y_s
    """
    
    AUCs_target = []
    AUCs_proxy = []
    # Set up target Y,proxy Y_s and features X 
    Y = DF[target_column_name]
    Y_s = DF[proxy_column_name] 
    X = DF.drop([proxy_column_name,target_column_name],axis = 1)
    
    # Set up Crossvalidation
    train_ix, test_ix =splitTestTrainIndices(working_df , target_column_name, proxy_column_name, min_frac_positive)
    X_train, X_test = X.ix[train_ix], X.ix[test_ix]
    Y_train, Y_test = Y.ix[train_ix], Y.ix[test_ix]
    Y_s_train, Y_s_test = Y_s.ix[train_ix], Y_s.ix[test_ix]
    
    #make sure proxy and target have the same shape
    assert Y_s_train.shape == Y_train.shape
    assert Y_s_test.shape == Y_test.shape
    
    # init classifier instance
    clf = linear_model.LogisticRegression() #C=1e30

    # train on Y_s
    clf.fit(X_train, Y_s_train)

    # evaluate AUC for Y
    AUC_target = roc_auc_score(y_true = Y_test, y_score = clf.decision_function(X_test))
    
    # evaluate AUC for Y_s. as a check of how informative the model is
    AUC_proxy = roc_auc_score(y_true = Y_s_test, y_score = clf.decision_function(X_test))
    
    return AUC_target, AUC_proxy
   
```

```python
def meanAucWithProxy(num_of_trials = 100, min_frac_positive = 0.78):
    """ Function evaluated the mean and standard deviation of AUC for both the predicted target variable 
    and predicted proxy variable
        Args:
            num_of_trials: int. the number of times the logistic regression will be run to calculate AUC.
            min_frac_positive: float. the mean fraction of overall positives we want the training DF to contain. default = 0.75.
        Returns:
    """
    AUCs_target = []
    AUCs_proxy = []
    
    for i in xrange(num_of_trials):
        
        # get the AUC for Target and Proxy for each instance of logistics regression
        AUC_target, AUC_proxy = proxyLogisticRegression(working_df,'Bin 2013', 'Bin 2014', min_frac_positive)
        
        # calculate the mean and std of AUC for target and proxy
        AUCs_target.append(AUC_target)
        AUCs_proxy.append(AUC_proxy)
        mean = np.mean(AUCs_target), np.mean(AUCs_proxy)
        std = np.std(AUCs_target), np.std(AUCs_proxy)
        
    return mean, std

meanAucWithProxy()
```

```python
def AucVariationWithPositives(frac_range):
    """ this function tests how the mean AUC changes for both target and proxy as 
    the result of changing the minimun fraction of positives present in the train sample.
    
    Args:
        frac_range: list of minimum fractions.
        
    Returns:
        list_AUC_target: list of target AUC. a member corresponding for each element in frac_range
        list_AUC_proxy: list of proxy AUC. a member corresponding for each element in frac_range
    
    """
    list_AUC_target =[]
    list_AUC_proxy = []
    for f in frac_range:
        print "testing for fraction: ", f
        mean, std = meanAucWithProxy(num_of_trials = 200, min_frac_positive = f)
        mean_AUC_target, mean_AUC_proxy  = mean
        list_AUC_target.append(mean_AUC_target)
        list_AUC_proxy.append(mean_AUC_proxy)
    return list_AUC_target, list_AUC_proxy

frac_range = [0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85]

list_AUC_target, list_AUC_proxy = AucVariationWithPositives(frac_range)
```
