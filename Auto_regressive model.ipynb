```python

# This function is based on an example from http://scikit-learn.org/0.11/auto_examples/linear_model/plot_ols.html

def scoresSelectiveLinearoReg(DF, measures, auto_regressive = True):
    """  function takes in a pandas DF, extracts relevent columns for auto regression and runs a multiple linear 
    regression model.
    
    Args:
        DF : a pandas dataframe containing at least columns for "Score 2014" and one extra year.
        
        measures: list of measures on which to run the linear regression. 
            For auto regressive model input measures such as: ['Score 2012', 'Score 2013'].
            
        auto_regressive = boolean. if True, performs uses only columns of scores from different years.
    
    Returns:
        mean_accuracy: float. mean accuracy of classifier.
        
    """
    if auto_regressive == True:
        scores_only_df  = DF.filter(regex="Score")
        scores_no_missing = scores_only_df.dropna()
    else:
        scores_no_missing = DF.dropna()
    
    
    Y = scores_no_missing['Score 2014']
    X = scores_no_missing[measures]
    
    #split into test and train DFs
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
    
    regr = linear_model.LinearRegression()
    regr.fit(X, Y)
    
    print 'Coefficients:\n', regr.coef_
    
    print("Residual sum of squares: %.2f"
      % np.mean((regr.predict(X_test) - Y_test) ** 2))
    
    # Explained variance score: 1 is perfect prediction
    mean_accuracy = regr.score(X_test, Y_test)
    print('Variance score: %.2f' % mean_accuracy)
    print "------------"
 
    return mean_accuracy

scoresSelectiveLinearoReg(df_joined, ['Score 2012', 'Score 2013'])
scoresSelectiveLinearoReg(df_joined, ['Score 2012'])
scoresSelectiveLinearoReg(df_joined, ['Score 2013'])
```

```python
# this should be switched out for whatever more reasonable function we come up with.
def fakeBucketTargetScores(target_series):
    # TODO make real bucket function
    return target_series.astype(float).map(lambda score: 0 if score < 1 else 1)
```

```python
def autoRegBucketedScores(DF, measures, bucketing_function, auto_regressive = True):
    """  function takes in a pandas DF, extracts relevent columns for auto regression and runs a multiple linear 
    regression model.
    
    Args:
        DF : a pandas dataframe containing at least columns for "Score 2014" and one extra year.
        
        measures: list of measures on which to run the linear regression. 
            For auto regressive model input measures such as: ['Score 2012', 'Score 2013'].

        bucketing_function: a function with which to perform the bucketing on the target variable.
        
        auto_regressive = boolean. if True, performs uses only columns of scores from different years.
    
    Returns:
        auc: float. mean accuracy of classifier.
        also plots the roc curve.
        
    """
    
    # only keep score columns in DF.
    scores_only_df  = DF.filter(regex="Score")
    scores_no_missing = scores_only_df.dropna()
    
    # bucket data
    s1 = scores_no_missing['Score 2014']
    scores_no_missing['bucketed_2014'] = bucketing_function(s1)


    Y = scores_no_missing['bucketed_2014']
    X = scores_no_missing[measures]

    #split into test and train DFs
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)

    #perform logistic regression 
    lr_classifier = linear_model.LogisticRegression(C=1e30)
    lr_classifier.fit(X_train, Y_train)
    lr_scores = lr_classifier.decision_function(X_test)

    fpr, tpr, thresholds = roc_curve(Y_test, lr_scores)
    roc_auc =  auc(fpr, tpr)
    print 'logistic regression AUC:', roc_auc
    plt.plot(fpr, tpr, label= ' (AUC = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.title('ROC')
    plt.legend(loc="lower right")
    plt.show()

    
autoRegBucketedScores(df_joined, measures = ['Score 2012', 'Score 2013'], bucketing_function = fakeBucketTargetScores)
```
