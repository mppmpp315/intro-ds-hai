```python
import sys
sys.path.append('src/')
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import sklearn.metrics as metrics
from sklearn import feature_selection
from sklearn import linear_model

import classifier
reload(classifier)
import data_utils
import binning_utils
import merge_data
reload(merge_data)
```

```python
all_label = merge_data.mergeAllTheThings(hai_2014_binning_func=binning_utils.binByLabel, 
                                         hai_2013_binning_func=binning_utils.binByScore4, 
                                         hai_2012_binning_func=binning_utils.binByScore4)
all_bin_1 = merge_data.mergeAllTheThings(hai_2014_binning_func=binning_utils.binByScore1, 
                                         hai_2013_binning_func=binning_utils.binByScore4, 
                                         hai_2012_binning_func=binning_utils.binByScore4)
all_bin_2 = merge_data.mergeAllTheThings(hai_2014_binning_func=binning_utils.binByScore2, 
                                         hai_2013_binning_func=binning_utils.binByScore4, 
                                         hai_2012_binning_func=binning_utils.binByScore4)
all_bin_3 = merge_data.mergeAllTheThings(hai_2014_binning_func=binning_utils.binByScore3, 
                                         hai_2013_binning_func=binning_utils.binByScore4, 
                                         hai_2012_binning_func=binning_utils.binByScore4)
```

```python
# Splitting the data into training and testing rows. 
# Even though I'm using the hai_bin_label data to do this, it will work for all the datasets since 
# they all have the same index.
train_ix, test_ix = data_utils.splitTestTrainIndices(all_label, 'Bin 2014', train_size=0.8)
```

```python
# 1) All available data, including the unbinned 2012 and 2013 HAI data, unbinned, L2 regularization

l2_classifier = classifier.UnfittedLogisticRegression('roc_auc', regularization_type='l2')

classifier.GraphROCs(all_label, all_bin_1, all_bin_2, all_bin_3, train_ix, test_ix, l2_classifier)
```

```python
# Takeaways:
# Not any better than autoregressive! Slightly worse, in fact.
```

```python
# 2) All available data, including the unbinned 2012 and 2013 HAI data, unbinned, L1 regularization
# NOTE: only graphed bin 3 because this takes a long time.

l1_classifier = classifier.UnfittedLogisticRegression('roc_auc', regularization_type='l1')
classifier.GraphROCs(all_label, all_bin_1, all_bin_2, all_bin_3, train_ix, test_ix, l1_classifier)
```

```python
# 3) All available data, paired down by recursive feature elimination.
# NOTE: only graphed bin 3 because this takes a long time.

# Using default logisitic regression model (C=1, l2 regularization), and 3 folds when deciding if a column is useful.
rfecv_classifier = feature_selection.RFECV(linear_model.LogisticRegression(), step=1, cv=3)
classifier.GraphROCs(all_label, all_bin_1, all_bin_2, all_bin_3, train_ix, test_ix, rfecv_classifier)
```

```python
# Takeaways:
# L1 regularization had no effect compared to L2 regularization.
```
