```python
# Build a Logistic Regression Based on
# 1) Just the 2013 HAI data, all three binning functions.
# 2) The 2012 and 2013 HAI data, all three binning functions.
# 3) The 2013 HAI data, unbinned.
# 4) The 2012 and 2013 HAI data, unbinned.

import sys
sys.path.append('src/')
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import metrics

import classifier
import data_utils
import binning_utils
import merge_data
```

```python
# Prepare all datasets
hai_bin_label = merge_data.mergeHAITables(hai_2014_binning_func=binning_utils.binByLabel, 
                                          hai_2013_binning_func=binning_utils.binByLabel, 
                                          hai_2012_binning_func=binning_utils.binByLabel)
hai_bin_1 = merge_data.mergeHAITables(hai_2014_binning_func=binning_utils.binByScore1, 
                                      hai_2013_binning_func=binning_utils.binByScore1, 
                                      hai_2012_binning_func=binning_utils.binByScore1)
hai_bin_2 = merge_data.mergeHAITables(hai_2014_binning_func=binning_utils.binByScore2, 
                                      hai_2013_binning_func=binning_utils.binByScore2, 
                                      hai_2012_binning_func=binning_utils.binByScore2)
hai_bin_3 = merge_data.mergeHAITables(hai_2014_binning_func=binning_utils.binByScore3, 
                                      hai_2013_binning_func=binning_utils.binByScore3, 
                                      hai_2012_binning_func=binning_utils.binByScore3)
```

```python
# Splitting the data into training and testing rows. 
# Even though I'm using the hai_bin_label data to do this, it will work for all the datasets since 
# they all have the same index.
train_ix, test_ix = data_utils.splitTestTrainIndices(hai_bin_label, 'Bin 2014', train_size=0.8)
```

```python
# 1) 2013 data
hai_2013_bin_label = pd.concat([hai_bin_label['Bin 2014'], hai_bin_label['Bin 2013']], axis=1)
hai_2013_bin_1 = pd.concat([hai_bin_1['Bin 2014'], hai_bin_1['Bin 2013']], axis=1)
hai_2013_bin_2 = pd.concat([hai_bin_2['Bin 2014'], hai_bin_2['Bin 2013']], axis=1)
hai_2013_bin_3 = pd.concat([hai_bin_3['Bin 2014'], hai_bin_3['Bin 2013']], axis=1)
```

```python
def GetLogisticROCData(data, train_ix, test_ix):
    X, Y = data_utils.splitTrainAndTarget(data)
    X_train, X_test = X.ix[train_ix], X.ix[test_ix]
    Y_train, Y_test = Y.ix[train_ix], Y.ix[test_ix]
    print '-- Number of positives: Y train', sum(Y_train)
    print '-- Number of positives: Y test', sum(Y_test)
    lr_classifier = classifier.optimizeLogisticRegression(X_train, Y_train, 'roc_auc')
    fpr, tpr, thresholds = metrics.roc_curve(Y_test, lr_classifier.decision_function(X_test))
    print '-- Number of ROC threshold values', len(thresholds)    
    return fpr, tpr, metrics.auc(fpr, tpr)
```

```python
# Graph AUC curves of all four binning functions
def GraphROCs(label_data, bin_1_data, bin_2_data, bin_3_data, train_ix, test_ix):
    # Train classifiers, and get ROC curve data
    print 'Bin: label'
    lr_fpr_label, lr_tpr_label, lr_auc_label = GetLogisticROCData(label_data, train_ix, test_ix)
    print 'Bin: 1'    
    lr_fpr_1, lr_tpr_1, lr_auc_1 = GetLogisticROCData(bin_1_data, train_ix, test_ix)
    print 'Bin: 2'    
    lr_fpr_2, lr_tpr_2, lr_auc_2 = GetLogisticROCData(bin_2_data, train_ix, test_ix)
    print 'Bin: 3'    
    lr_fpr_3, lr_tpr_3, lr_auc_3 = GetLogisticROCData(bin_3_data, train_ix, test_ix)    
    
    fig, ax = plt.subplots()
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title('ROC')    
    legends = []

    plt.plot(lr_fpr_label, lr_tpr_label)
    legends.append('LR, Original Labels (AUC = %0.2f)' % (lr_auc_label))
    plt.plot(lr_fpr_1, lr_tpr_1)
    legends.append('LR, Binning 1 (AUC = %0.2f)' % (lr_auc_1))
    plt.plot(lr_fpr_2, lr_tpr_2)
    legends.append('LR, Binning 2 (AUC = %0.2f)' % (lr_auc_2))
    plt.plot(lr_fpr_3, lr_tpr_3)
    legends.append('LR, Binning 3 (AUC = %0.2f)' % (lr_auc_3))

    # Straight line
    diagonal_fpr = [0,1]
    diagonal_tpr = [0,1]
    plt.plot(diagonal_fpr, diagonal_tpr)
    legends.append('Random Guess (AUC = %0.2f)' % (metrics.auc(diagonal_fpr, diagonal_tpr)))
    
    plt.legend(legends, title='Classifier Type', loc='lower right')
    plt.show()

GraphROCs(hai_2013_bin_label, hai_2013_bin_1, hai_2013_bin_2, hai_2013_bin_3, train_ix, test_ix)
```

```python
# Takeaways:
# These curves change quite a bit when you regenerate them, especially the original labels curve.
# There is only one feature, and it is binary. So the model consists entirely on one coefficient. 
# Therefore it can only do three things:
# 1) Classify everything negatively.
# 2) If the 2013 score is 1, classify positively.
# 3) Classify everything positively.

# Therefore, this model is a 'lazy weatherman' model. It just predicts that value from last year.
# The variability comes from the random differences in the train/test split.
```

```python
# 2) 2012 and 2013 data

hai_both_bin_label = pd.concat([hai_bin_label['Bin 2014'], hai_bin_label['Bin 2013'], hai_bin_label['Bin 2012']], axis=1)
hai_both_bin_1 = pd.concat([hai_bin_1['Bin 2014'], hai_bin_1['Bin 2013'], hai_bin_1['Bin 2012']], axis=1)
hai_both_bin_2 = pd.concat([hai_bin_2['Bin 2014'], hai_bin_2['Bin 2013'], hai_bin_2['Bin 2012']], axis=1)
hai_both_bin_3 = pd.concat([hai_bin_3['Bin 2014'], hai_bin_3['Bin 2013'], hai_bin_3['Bin 2012']], axis=1)
```

```python
GraphROCs(hai_both_bin_label, hai_both_bin_1, hai_both_bin_2, hai_both_bin_3, train_ix, test_ix)
```

```python
# Takeaways:
# This model is very similar to the last one. Now that it has two years, it has 4 (2*2) different feature vector possibilities.
# The weights for those two features determine the outcome. As a result, there is more preditive power in the model, and
# the AUC values are slightly higher (but not by much).
```

```python
# Load non-binned data   
hai_nobin_label = merge_data.mergeHAITables(hai_2014_binning_func=binning_utils.binByLabel, 
                                            hai_2013_binning_func=binning_utils.binByScore4, 
                                            hai_2012_binning_func=binning_utils.binByScore4)
hai_nobin_1 = merge_data.mergeHAITables(hai_2014_binning_func=binning_utils.binByScore1, 
                                      hai_2013_binning_func=binning_utils.binByScore4, 
                                      hai_2012_binning_func=binning_utils.binByScore4)
hai_nobin_2 = merge_data.mergeHAITables(hai_2014_binning_func=binning_utils.binByScore2, 
                                      hai_2013_binning_func=binning_utils.binByScore4, 
                                      hai_2012_binning_func=binning_utils.binByScore4)
hai_nobin_3 = merge_data.mergeHAITables(hai_2014_binning_func=binning_utils.binByScore3, 
                                      hai_2013_binning_func=binning_utils.binByScore4, 
                                      hai_2012_binning_func=binning_utils.binByScore4)
```

```python
# 3) 2013 data, unbinned.
hai_2013_nobin_label = pd.concat([hai_nobin_label['Bin 2014'], hai_nobin_label['Bin 2013']], axis=1)
hai_2013_nobin_1 = pd.concat([hai_nobin_1['Bin 2014'], hai_nobin_1['Bin 2013']], axis=1)
hai_2013_nobin_2 = pd.concat([hai_nobin_2['Bin 2014'], hai_nobin_2['Bin 2013']], axis=1)
hai_2013_nobin_3 = pd.concat([hai_nobin_3['Bin 2014'], hai_nobin_3['Bin 2013']], axis=1)
```

```python
GraphROCs(hai_2013_nobin_label, hai_2013_nobin_1, hai_2013_nobin_2, hai_2013_nobin_3, train_ix, test_ix)
```

```python
# Takeaways:
# Unbinning the feature HAI scores gives much more stability to the model. Since the one feature isn't binary,
# a small change in the threshold no longer results in everything being labeled as positive.
# Although the "Original Labels" curve has nearly the greatest AUC, it has the highest variability.
# That is, if you regenerate this graph with a slightly different train/test split, that curve changes dramatically.
# This happens because there are so few positive labels in that set, so small changes in the threshold can affect a large
# fraction of the test set.
# In contrast, the other binning functions have many more positive examples, so small changes in the threshold don't affect
# a large fraction of the test set.
```

```python
# 4) 2012 and 2013 data, unbinned.
hai_both_nobin_label = pd.concat([hai_nobin_label['Bin 2014'], hai_nobin_label['Bin 2013'], hai_nobin_label['Bin 2012']], axis=1)
hai_both_nobin_1 = pd.concat([hai_nobin_1['Bin 2014'], hai_nobin_1['Bin 2013'], hai_nobin_1['Bin 2012']], axis=1)
hai_both_nobin_2 = pd.concat([hai_nobin_2['Bin 2014'], hai_nobin_2['Bin 2013'], hai_nobin_2['Bin 2012']], axis=1)
hai_both_nobin_3 = pd.concat([hai_nobin_3['Bin 2014'], hai_nobin_3['Bin 2013'], hai_nobin_3['Bin 2012']], axis=1)
```

```python
GraphROCs(hai_both_nobin_label, hai_both_nobin_1, hai_both_nobin_2, hai_both_nobin_3, train_ix, test_ix)
```

```python
# Takeaways:
# No notable gains from adding the 2012 data. It does about as well as it did before. 
# The Original Label curve is still highly variable. Adding the 2012 didn't stabilize it.
```
