```python
import pandas as pd
import numpy as np
from math import * 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def dataload(file_name):
    DF = pd.read_csv(file_name)
    DF = DF.set_index(DF['Provider ID'])
    DF = DF.drop('Provider ID', 1)
    return DF
```

```python
file_1 = 'data/Healthcare Associated Infections - Hospital.csv'
file_2 = 'data/Medicare Hospital Spending per Patient - Hospital.csv'
file_3 = 'data/Medicare Volume - Hospital.csv'

Infections = dataload(file_1)
Spending = dataload(file_2)
Volume = dataload(file_3)
```

```python
#pear down repeating before join

def shrink(DF):
    to_drop = []
    for name in DF.columns:
        #if in dataframe, drop general information columns
        if name in ['Hospital Name','Address', 'City', 'State', 'ZIP Code', 'County Name', 'Phone Number']:
            to_drop.append(name)
    DF = DF.drop( to_drop,1)
    return DF

V1 = shrink(Volume)
S1 = shrink(Spending)
```

```python
'''Volume
#134245 rows × 12 columns
Spending

#3320 rows × 13 columns

Infections
#168588 rows × 14 columns'''

def dataframeJoin(DF_infection, DFs_to_merge):

    for count, DF in enumerate(DFs_to_merge):
        print "DF", count
        
        joined = DF_infection.join(DF, how='left', rsuffix='_%s' %count)
    return joined
```

```python
#example of joining
Joined  = dataframeJoin(Infections, [Volume, Spending])
```

```python
infection_scores = Joined.Score[Joined.Score != 'Not Available']
```
